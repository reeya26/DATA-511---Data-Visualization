```python
#Loading libraries for Data Cleaning and Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#Loading the World Indicators Dataset
df = pd.read_csv("world indicators exercise.csv")
```

```python
#Q1 - Removing rows where Country Name is null

#what data is present in the Country Name column?
df['Country Name'].unique()
```

```python
#Removing the invalid rows from the Data, using the Country Name column

df_clean = df[(df['Country Name'] != "Data from database: World Development Indicators") & 
              (df['Country Name'] != "Last Updated: 04/24/2019")]

#Cleaned df after removing invalid rows
print(df_clean['Country Name'].unique())
```

```python
#Q2 - Filter the data to just the United States and United Kingdom


#df_filter = df_clean.loc[(df['Country Name'] == "United States") & (df['Country Name'] == "United Kingdom")]
#(df_clean)


df_filter = df_clean.loc[(df['Country Name'] == "United States") | (df['Country Name'] == "United Kingdom")] 
df_filter
```

```python
#Q3 - Drop  the Series Code column

df_drop = df_filter.drop('Series Code',axis=1)
df_drop.columns
```

```python
#Q4 - Pivot all years of data into a single column

#Rearranging the data to transform the "Date" columns as variables in Rows
df_pivot = pd.melt(df_drop,id_vars=['Country Name', 'Country Code', 'Series Name'], value_vars=['1960 [YR1960]',
       '1961 [YR1961]', '1962 [YR1962]', '1963 [YR1963]', '1964 [YR1964]',
       '1965 [YR1965]', '1966 [YR1966]', '1967 [YR1967]', '1968 [YR1968]',
       '1969 [YR1969]', '1970 [YR1970]', '1971 [YR1971]', '1972 [YR1972]',
       '1973 [YR1973]', '1974 [YR1974]', '1975 [YR1975]', '1976 [YR1976]',
       '1977 [YR1977]', '1978 [YR1978]', '1979 [YR1979]', '1980 [YR1980]',
       '1981 [YR1981]', '1982 [YR1982]', '1983 [YR1983]', '1984 [YR1984]',
       '1985 [YR1985]', '1986 [YR1986]', '1987 [YR1987]', '1988 [YR1988]',
       '1989 [YR1989]', '1990 [YR1990]', '1991 [YR1991]', '1992 [YR1992]',
       '1993 [YR1993]', '1994 [YR1994]', '1995 [YR1995]', '1996 [YR1996]',
       '1997 [YR1997]', '1998 [YR1998]', '1999 [YR1999]', '2000 [YR2000]',
       '2001 [YR2001]', '2002 [YR2002]', '2003 [YR2003]', '2004 [YR2004]',
       '2005 [YR2005]', '2006 [YR2006]', '2007 [YR2007]', '2008 [YR2008]',
       '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]',
       '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]',
       '2017 [YR2017]', '2018 [YR2018]'],var_name='Year', value_name='Data')


df_pivot['Year'] = df_pivot['Year'].astype(str).str[0:4]

#Removing all rows where Data column is null or undefined

df_pivoted = df_pivot[(df_pivot['Data'] != "..")]

print(df_pivoted)
```

```python
#Q5 - Un-Pivot all indicator names from rows into columns



df_transformed = pd.pivot_table(df_pivoted, values='Data', index=['Country Name','Country Code','Year'],
                     columns=['Series Name'], aggfunc=np.sum)
df_transformed
```

```python
# Q6 - Drop all but the following indicators: CO2 emissions (kt), Population total,  Agricultural land (sq. km)


mask = df_pivoted['Series Name'].isin(['CO2 emissions (kt)', 'Population, total', 'Agricultural land (sq. km)'])
table = df_pivoted[mask].pivot_table(
        values='Data', 
        index=['Country Name','Country Code','Year'],
        columns=['Series Name'], 
        aggfunc=np.sum, 
    )

table
```

```python
# Q7 - Create a derived column called CO2 emissions (kt per person)
#      using the calculation: CO2 emissions(kt) / Population total


#table['CO2 emissions (kt per person)'] = (table['CO2 emissions (kt)']) / (table['Population, total'])

#table['CO2 emissions (kt)'].astype(str).astype(int)
#table['Population, total'].astype(str).astype(int)

table['CO2 emissions (kt)'] = pd.to_numeric(table['CO2 emissions (kt)'])
table['Population, total'] = pd.to_numeric(table['Population, total'])

table['CO2 emissions (kt per person)'] = (table['CO2 emissions (kt)']) / (table['Population, total'])
table
```

```python
# Final ouput is saved to a Xlsx File

table.to_excel("Assignment3.xlsx")
```
